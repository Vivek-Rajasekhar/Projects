```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('encoded_data_5000.csv')
df
```

```python
df.drop('Unnamed: 0', axis=1,inplace=True)
df
```

```python
X = df.drop('Burnout Risk', axis=1)
y = df['Burnout Risk']

X_train, x_test,y_train, y_test = train_test_split(X,y,train_size=.8)
```

```python
regg = LogisticRegression()
regg.fit(X_train, y_train)
```

```python
regg.predict(x_test)
```

```python
regg.score(x_test, y_test)
```

```python
new_data = pd.DataFrame({
    'Stress Level': ['Moderate', 'High'],
    'Mood Rating': ['Positive', 'Negative'],
    'Sleep Quality': ['Good', 'Poor'],
    'Energy Level': ['High', 'Low'],
    'Work Hours Per Day': ['Normal', 'Overworking'],
    'Job Satisfaction': ['Very Satisfied', 'Dissatisfied'],
    'Workload Intensity': ['Moderate', 'High'],
    'Break Frequency': ['Moderately Balanced', 'Low Balance'],
    'Overtime Frequency': ['Occasional', 'Constant'],
    'Role Clarity': ['Somewhat Clear', 'Unclear'],
    'Autonomy in Work': ['Moderate', 'Low']
})
categories_order = [
    ['Low', 'Moderate', 'High'],             # Stress Level
    ['Negative', 'Neutral', 'Positive'],     # Mood Rating
    ['Poor', 'Average', 'Good'],             # Sleep Quality
    ['Low', 'Moderate', 'High'],             # Energy Level
    ['Low', 'Normal', 'Extended', 'Overworking'],  # Work Hours Per Day
    ['Dissatisfied', 'Neutral', 'Very Satisfied'], # Job Satisfaction
    ['Low', 'Moderate', 'High'],             # Workload Intensity
    ['Low Balance', 'Moderately Balanced', 'Balanced'],  # Break Frequency
    ['None', 'Occasional', 'Frequent', 'Constant'],      # Overtime Frequency
    ['Unclear', 'Somewhat Clear', 'Very Clear'],         # Role Clarity
    ['Low', 'Moderate', 'High']              # Autonomy in Work
]
ordi = OrdinalEncoder(categories=categories_order)

columns_to_encode = [
    'Stress Level', 'Mood Rating', 'Sleep Quality', 'Energy Level', 'Work Hours Per Day',
    'Job Satisfaction', 'Workload Intensity', 'Break Frequency', 'Overtime Frequency',
    'Role Clarity', 'Autonomy in Work'
]
# Encode and predict
new_data[columns_to_encode] = ordi.fit_transform(new_data[columns_to_encode])
predictions = regg.predict(new_data)

print("Predicted Outputs:", predictions)
```
